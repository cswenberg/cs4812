```python
from qiskit import *
from math import pi
```

```python
print("PS3, problem 6")
print("a)")

qc = QuantumCircuit(3)

qc.h(2)
qc.cx(1,2)
qc.tdg(2)
qc.cx(0,2)
qc.t(2)
qc.cx(1,2)
qc.tdg(2)
qc.cx(0,2)
qc.tdg(1)
qc.tdg(2)
qc.cx(0,1)
qc.h(2)
qc.tdg(1)
qc.cx(0,1)
qc.t(0)
qc.s(1)

print(qc.draw(output='text'))

print("\nb)")

qc = QuantumCircuit(3)

qc.ry(pi/4,2)
qc.cx(1,2)
qc.ry(pi/4,2)
qc.cx(0,2)
qc.ry(-pi/4,2)
qc.cx(1,2)
qc.ry(-pi/4,2)

print(qc.draw(output='text'))
```

```python
print("PS4, problem 1")
print("a)")
print("(i)")

qc = QuantumCircuit(2)

qc.ry(pi/2,0)
qc.ry(pi/2,1)
qc.x(1)
qc.cz(0,1)
qc.x(1)
qc.ry(pi/2,0)

print(qc.draw(output='text'))

print("(ii)")

qc = QuantumCircuit(2)

qc.h(0)
qc.cx(0,1)

print(qc.draw(output='text'))

print("\nb)")
print("(i)")

qc = QuantumCircuit(3)

qc.ry(pi/2,0)
qc.ry(pi/2,1)
qc.ry(pi/2,2)
qc.x(1)
qc.cz(0,1)
qc.cz(2,1)
qc.x(1)
qc.ry(pi/2,0)
qc.ry(pi/2,2)

print(qc.draw(output='text'))

print("(ii)")

qc = QuantumCircuit(3)

qc.ry(pi/2,0)
qc.ry(pi/2,1)
qc.ry(pi/2,2)
qc.iswap(0,1)
qc.iswap(1,2)
qc.rx(-pi/2,0)
qc.rx(-pi/2,1)

print(qc.draw(output='text'))

print("(iii)")

qc = QuantumCircuit(3)

qc.h(0)
qc.cx(0,1)
qc.cx(1,2)

print(qc.draw(output='text'))
```

```python
print("PS4, problem 2")
print("a) with taking 0.6π as an approximation for alpha=arccos(-1/3)")

qc = QuantumCircuit(3)

qc.ry(0.6*pi, 0)
qc.cx(0,1)
qc.ch(1,2)
qc.cx(2,0)
qc.x(1)

print(qc.draw(output='text'))
```

```python
print("PS6, problem 2")
print("a)")

q = QuantumRegister(4)
c = ClassicalRegister(4)
qc = QuantumCircuit(q, c)

qc.h(0)
qc.cx(0,2)
qc.cx(0,3)
qc.h(0)
qc.cx(2,1)
qc.cx(3,1)
qc.measure(q[0:2],c[0:2])

print(qc.draw(output='text'))
```

```python
print("PS6, problem 6")
print("A.")

q = QuantumRegister(4)
c = ClassicalRegister(4)
qc = QuantumCircuit(q, c)

qc.h(0)
qc.cz(0,2)
qc.cz(0,3)
qc.h(0)
qc.h(3)
qc.cx(3,2)
qc.h(3)
qc.measure([q[0], q[3]],[c[0],c[3]])

print(qc.draw(output='text'))

print("B.")

qc = QuantumCircuit(3)

qc.cx(2,1)
qc.cx(1,0)
qc.t(0)
qc.tdg(1)
qc.t(2)
qc.cx(2,1)
qc.cx(1,0)
qc.tdg(0)
qc.cx(2,0)

print(qc.draw(output='text'))

print("C.")

q = QuantumRegister(6)
c = ClassicalRegister(6)
qc = QuantumCircuit(q, c)

qc.h(0)
qc.cz(0,1)
qc.cz(0,2)
qc.cz(0,3)
qc.cz(0,4)
qc.cz(0,5)
qc.h(0)
qc.h(5)
qc.cx(5,4)
qc.h(5)
qc.measure([q[0], q[5]],[c[0],c[5]])

print(qc.draw(output='text'))
```
